```python
# SET with mathematical ooperation
```

```python
myset1 = {2,5,6,3,8,1,19}
myset2 = {10,20,1,23,3,7,8}
myset3 = {5,8,9,3,20}
myset4 = {8,9,5,3}
myset5 = {10,20,1,23,3,7,8}
```

```python
# Union
print(myset1.union(myset2))
```

```python
# Intersection
print(myset1.intersection(myset2))
```

```python
# difference
print(myset1.difference(myset2)) # return values present in myset1 but not in myset2
print(myset2.difference(myset1)) # returns values present in myset2 but not in myset1
```

```python
# update the set with calculated values
myset3.intersection_update(myset1)
print(myset3)  # myset2 will be having values of intersection between myset1 and myset3

myset4.difference_update(myset5) # myset4 will have values of difference of values of sets i.e. value in myset4 but not in myset5
print(myset4)
```

```python
# symmetric_difference (if you wirite symmetric_difference_update then value will be updated)
print(myset2.symmetric_difference(myset1)) # it will ignore values which are present in both sets
```

```python
# copy
myset6 = myset1.copy()
print(myset6)
```

```python
# isdisjoint --> returns true if intersection if empty
x = {4,6}
y = {5,9}
print(x.isdisjoint(y))
a = {3,5}
b = {3,7}
print(a.isdisjoint(b))
```

```python
# issubset
n = {2,4,6}
m = {6,2}
print(m.issubset(n))
print(n.issubset(m))
```

```python
# issuperset
n = {2,4,6}
m = {6,2}
print(n.issuperset(m))
print(m.issuperset(n))
```
