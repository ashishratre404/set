Creation of set

2. x = set() : for empty set
3. x = set(iterable) : set with values

```python
a = {5,8,3}
print(a)
```

```python
b = {7,'ashish', 8, 9, 'riya'}
print(b)
```

```python
# we cant contain List because it is mutable
c = {5,6,[4,9]}
print(c)
```

```python
# we can contain tuple(immutable)
d = {'ashish', 8, 9, (4,7,9)}
print(d)
```

```python
# we cant have duplicate
e = {3,5,3,5,3,9} # it count only once, doesnt care how many times you have given same value, it will be printed only once
print(e)
```

Deletion
1. discard(value)
2. remove(value)
3. pop()
4. clear()

```python
set2 = {3,5,8,6,2}
```

```python
set2.discard(2)
print(set2)
```

```python
set2.remove(8)
print(set2)
```

```python
g = set2.pop() # deletes any random position
print(g)
print(set2)
```

```python
set2.clear()
print(set2)
```

Addition
1. add(value)
2. update([value1, value2,....])

```python
set3 = {4,6,7,2}
print(set3)
```

```python
# using add  (for adding one value at a time)
set3.add(9)
print(set3)
```

```python
# update(iterable..)  (for multiple values)
set3.update([11,54,8])
print(set3)
```

```python
# we can also pass multiple iterables
set3.update([33,77],{66,99,88}) #set is also iterable
print(set3)
```
